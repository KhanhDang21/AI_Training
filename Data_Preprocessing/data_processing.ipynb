```python
import pandas as pd
import numpy as np
```

1. Hiển thị toàn bộ dữ liệu

```python
data = pd.read_csv('P4AI_BT1.csv')
data
```

2. Điền các giá trị bị thiếu của dữ liệu theo phương pháp trung bình với các trường có giá trị số và theo phương pháp tần số cao nhất với giá trị rời rạc

- Phương pháp trung bình với các trường có giá trị số:


```python
df1 = data.copy() 
# sao chép dữ liệu
```

```python
display(df1.isna().sum())
display(df1)
```

```python
# Thay giá trị NaN bằng các giá trị trung bình
df1['petal.length'].fillna(round(data['petal.length'].mean()), inplace = True)
df1['petal.width'].fillna(round(data['petal.width'].mean()), inplace = True)
df1['sepal.length'].fillna(round(data['sepal.length'].mean()), inplace = True)
df1['sepal.width'].fillna(round(data['sepal.width'].mean()), inplace = True)
display(df1)
```

- Phương pháp tần số cao nhất với giá trị rời rạc

```python
# Thay giá trị NaN bằng giá trị xuất hiện với tần số lớn nhất
df1['variety'].fillna(df1['variety'].mode()[0], inplace = True)
display(df1)
```

- Kiểm tra lại


```python
df1.isna().sum()
```

3. Hiển thị các bản ghi có sepal.length > 5 và sepal.width > 3

```python
filltered_data = df1[(df1['sepal.length'] > 5) & (df1['sepal.width'] > 3)]
display(filltered_data)
```

4. Chuẩn hóa dữ liệu theo min max

- Chuẩn hóa 4 thuộc tính là sepal.length, sepal.width, petal.length, petal.width

```python
df2 = df1.copy()
```

```python
sepal_length_min = df2['sepal.length'].min()
sepal_length_max = df2['sepal.length'].max()

sepal_width_min = df2['sepal.width'].min()
sepal_width_max = df2['sepal.width'].max()

petal_length_min = df2['petal.length'].min()
petal_length_max = df2['petal.length'].max()

petal_width_min = df2['petal.width'].min()
petal_width_max = df2['petal.width'].max()
```

```python
# Thực hiện chuẩn hóa theo min max với 4 trường dữ liệu ta có
df2['sepal.length'] = (df2['sepal.length'] - sepal_length_min) / (sepal_length_max - sepal_length_min)
df2['sepal.width'] = (df2['sepal.width'] - sepal_width_min) / (sepal_width_max - sepal_width_min)
df2['petal.length'] = (df2['petal.length'] - petal_length_min) / (petal_length_max - petal_length_min)
df2['petal.width'] = (df2['petal.width'] - petal_width_min) / (petal_width_max - petal_width_min)
display(df2)
```

5. Liên tục hóa thuộc tính variety

```python
df3 = df2.copy()
```

Kiểm tra các thông tin trường variety

```python
df3['variety'].unique()
```

Đưa dữ liệu trường variety về dạng One-Hot Encoder

```python
dummies = pd.get_dummies(df3['variety'])
display(dummies)
```

Nối các trường dữ liệu ở dạng One-Hot với dạng dữ liệu ban đầu

```python
df3 = pd.concat([df3, dummies], axis = 1)
df3.drop('variety', axis = 1, inplace = True)
display(df3)
```

6. Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính variety dựa trên các thuộc tính khác

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Sao chép dữ liệu
df4 = df3.copy()
```

```python
# Chuyển hóa dữ liệu
df4['Setosa'].replace({"False": 0, "True": 1}, inplace = True)
df4['Versicolor'].replace({"False": 0, "True": 1}, inplace = True)
df4['Virginica'].replace({"False": 0, "True": 1}, inplace = True)
```

```python
# Tạo bộ test và bộ train
x = df4[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y = df4[['Setosa', 'Versicolor', 'Virginica']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# Huấn luyện dữ liệu
model = LinearRegression()  
model.fit(x_train, y_train)
```

```python
# Dự đoán kết quả
y_pred = model.predict(x_test)
```

```python
# Đánh giá mô hình và in ra độ chính xác
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

8. Rời rạc hóa thuộc tính sepal.length

```python
# Sao chép dữ liệu
df6 = df4.copy()
```

Rời rạc hóa thuộc tính sepal.length

```python
# Rời rạc hóa thuộc tính sepal.length
size = pd.cut(df4['sepal.length'],
    bins = [-0.1, 0.3, 0.4, 0.6, 0.8, 1.0],
    labels = ["Very low", "Low", "Medium", "High", "Very high"])
df6['Size'] = size
display(df6)
```

9. Phân cụm dữ liệu và loại bỏ ngoại lai 

```python
# Sao chép dữ liệu
df7 = df6.copy()
```

```python
from sklearn.cluster import KMeans, DBSCAN
```

Khởi tạo mô hình KMeans với 3 cụm

```python
kmeans = KMeans(n_clusters = 3, random_state = 42)
```

Phân cụm và gán nhãn cho từng cụm dữ liệu

```python
df7['cluster'] = kmeans.fit_predict(df7[['Setosa', 'Versicolor', 'Virginica']])
```

Sử dụng DBSCAN để loại bỏ ngoại lai

```python
dbscan = DBSCAN(eps = 0.3, min_samples = 2)
```

Gán nhãn ngoại lai

```python
df7['outlier'] = dbscan.fit_predict(df7[['Setosa', 'Versicolor', 'Virginica']])
```

Loại bỏ các điểm ngoại lai (nhãn là -1)

```python
df_no_outlier = df7[df7['outlier'] != -1]
display(df_no_outlier)
```
